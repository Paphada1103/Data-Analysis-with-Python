## Bivariate Visualization Techniques

### 1. Bar Plot: Comparing Categories with Numerical Values

#### Example 1: Bar Plot for age by class

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset("titanic")

# Plotting a bar plot for 'age' by 'class'
sns.barplot(x="class", y="age", data=df)
```

#### Example 2: Bar Plot with Hue for sex

```python
# Plotting a bar plot for 'age' by 'class', with hue 'sex'
sns.barplot(x="class", y="age", hue="sex", data=df)
```

#### Example 3: Bar Plot with Sum of Values

```python
# Plotting a bar plot with sum of 'age' by 'class'
sns.barplot(x="class", y="age", estimator="sum", data=df)
```

### 2. Box Plot: Understanding Distribution and Outliers

#### Example 1: Box Plot for age by sex

```python
# Plotting a box plot for 'age' by 'sex'
sns.boxplot(x="sex", y="age", data=df)
```

#### Example 2: Box Plot with hue for class

```python
# Plotting a box plot for 'age' by 'sex', with hue 'class'
sns.boxplot(x="sex", y="age", hue="class", data=df)
```

### 3. Violin Plot: Combining Distribution and Box Plot

#### Example 1: Violin Plot for fare by sex

```python
# Plotting a violin plot for 'fare' by 'sex'
sns.violinplot(x="sex", y="fare", data=df)
```

### 4. Distribution Plot: Comparing Categories

```python
# Plotting distribution plots for 'age' by 'survived'
sns.displot(df[df["survived"] == 0]["age"], kind='kde')
sns.displot(df[df["survived"] == 1]["age"], kind='kde')
```
