## 1. Histogram: Visualizing Frequency Distribution


### Example 1: Histogram for price Column

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# Load the dataset
df = pd.read_csv("AB_NYC_2019.csv")

# Plotting a histogram for the 'price' column
sns.histplot(df["price"])
```

### Example 2: Filtering Values for a Clearer Distribution

```python
# Plotting a histogram for prices less than $500
sns.histplot(df[df["price"] < 500]["price"])
```

### Example 3: Histogram for number_of_reviews Column


```python
# Plotting a histogram for the 'number_of_reviews' column
sns.histplot(df[df["number_of_reviews"] < 100]["number_of_reviews"])
```

## 2. Distplot: Visualizing Probability Distribution

### Example 1: Distplot for price Column


```python
# Plotting a histplot for the 'price' column
sns.histplot(df["price"])
```

### Example 2: Binned Price Data

```python
# Binned value counts for the 'price' column
df["price"].value_counts(bins=20)
```

## 3. Box Plot: Identifying Outliers

### Example 1: Box Plot for price Column


```python
# Plotting a box plot for the 'price' column
sns.boxplot(df[df["price"] < 500]["price"])
```

### Example 2: Box Plot for reviews_per_month

```python
# Plotting a box plot for the 'reviews_per_month' column
sns.boxplot(df[df["reviews_per_month"] < 10]["reviews_per_month"])
```
