```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = sns.load_dataset("titanic")
df.head()
```

## 1. Scatter Plot for Numerical Data

```python
sns.scatterplot(x="age", y="fare", data=df)
```

## 2. Scatter Plot on the Tips Dataset

```python
df2 = sns.load_dataset("tips")
df2.head()

sns.scatterplot(x="total_bill", y="tip", data=df2)

# Adding Hue and Style
sns.scatterplot(x="total_bill", y="tip", hue="smoker", style="sex", data=df2)
```

## 3. Joint Plot for Numerical Data

```python
sns.jointplot(x="total_bill", y="tip",hue="smoker", data=df2)
```

#### Regression in Joint Plot

```python
sns.jointplot(x= "total_bill", y= "tip",data = df2, kind ="reg")
```

### Basic Line Plot

```python
sns.lineplot(x= "total_bill", y= "tip", data=df2)
```
