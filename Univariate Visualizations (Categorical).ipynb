# Visualizing Categorical Data

## Bar Graphs

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("AB_NYC_2019.csv")

# Plotting a bar graph for the 'neighbourhood_group' column
sns.countplot(x="neighbourhood_group", data=df)
```

```python
sns.countplot(data=df, x="neighbourhood_group", hue="room_type")
```

## Pie Charts

```python
df["room_type"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
df["neighbourhood_group"].value_counts().plot(kind="pie", autopct="%.2f")
```

# More Complex Visualizations


## Example 1: Bar Graph for survived Column


```python
df2 = sns.load_dataset("titanic")

# Plotting a count plot for survival, grouped by gender
sns.countplot(x="survived", data=df2)
```

## Example 2: Bar Graph for embark_town Column

```python
sns.countplot(x="embark_town", data=df2)
```

## Example 3: Pie Chart for class Column

```python
df2["class"].value_counts().plot(kind="pie", autopct="%.2f")
```

### Using Hue for More Insights

```python
sns.countplot(x="survived", data=df2, hue="sex")
```
