## 1. Bar Plot: Comparing Two Categorical Variables

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset("titanic")

# Bar plot for sex and age categorized by survival status
sns.barplot(x="sex", y="age", hue="survived", data=df)
```

## 2. Strip Plot: Understanding Data Distribution Across Categories


```python
# Strip plot for passenger class vs age, categorized by sex
sns.stripplot(x="class", y="age", hue="sex", data=df)
```

## 3. Heatmap: Analyzing Relationships in Categorical Data

```python
# Create a pivot table for survival counts in each class
x = pd.crosstab(df["class"], df["survived"])
print(x)
print('-'* 20)

# Plotting the heatmap
sns.heatmap(x, annot=True, fmt=".1f")
```

## 4. Stacked Bar Chart: Visualizing Multiple Categorical Variables

```python
df2 = sns.load_dataset("iris")
df2.head()

# Bar plot for the first 10 rows of the dataset
df2.head(10).plot(kind="bar")

# Stacked bar chart
df2.head(10).plot(kind="bar", stacked=True)
```

## 5. Pair Plot: Understanding Relationships Between Multiple Variables

```python
# Pair plot for the Iris dataset
sns.pairplot(df2)
```
